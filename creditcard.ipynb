**from the datasets we can directly remove some columns.. which are unique**

```python
train_df=train_df.drop(["first","last","street","dob","unix_time","zip","trans_date_trans_time","trans_num"],axis=1)
train_df
```

```python
test_df=test_df.drop(["first","last","street","dob","unix_time","zip","trans_date_trans_time","trans_num"],axis=1)
test_df
```

```python
train_df.dtypes
```

**train data  categorical conversion into numerical columns**

```python
from sklearn.preprocessing import LabelEncoder
for i,j in dict(train_df.dtypes).items():
    if j=='O':
        LE=LabelEncoder()
        train_df[i]=LE.fit_transform(train_df[i])
        
```

**test data categerocial data conversion into numerical data**

```python

from sklearn.preprocessing import LabelEncoder
for i,j in dict(test_df.dtypes).items():
    if j=='O':
        LE=LabelEncoder()
        test_df[i]=LE.fit_transform(test_df[i])
        
```

**developing the model**

```python
X_train=train_df.drop("is_fraud",axis=1)
y_train=train_df["is_fraud"]
X_test=test_df.drop("is_fraud",axis=1)
y_test=test_df["is_fraud"]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
DT.fit(X_train,y_train)
```

```python
y_pred=DT.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,confusion_matrix,ConfusionMatrixDisplay,roc_curve
acc_log= round(accuracy_score(y_test,y_pred)*100,2)
f1_log=round(f1_score(y_test,y_pred),2)
precision_log=round(precision_score(y_test,y_pred),2)
recall_log=round(recall_score(y_test,y_pred),2)
print("accuray is:",acc_log)
print("F1 is:",f1_log)
print("Precision is:",precision_log)
print("Recall is:",recall_log)
```

```python
cmt=confusion_matrix(y_test,y_pred)

disp=ConfusionMatrixDisplay(confusion_matrix=cmt,
                            display_labels = [False, True])
disp.plot()
plt.grid(False)
plt.show()


tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()
print("True negative:",tn)
print("False postive:",fp)
print("False negative:",fn)
print("True postive:",tp)
```

```python
y_dt_pred_prob=DT.predict_proba(X_test)[:,1]  
fpr,tpr,threshold=roc_curve(y_test,y_dt_pred_prob) 
plt.plot(fpr,tpr)
plt.show()
```
