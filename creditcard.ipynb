```python
#developing the model
```

```python
X_train=train_df.drop("is_fraud",axis=1)
y_train=train_df["is_fraud"]
X_test=test_df.drop("is_fraud",axis=1)
y_test=test_df["is_fraud"]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from collections import Counter
before_count=Counter(y_train)
before_count
```

```python
#leads to class imbalance problem
#so we use SMOTE
```

```python
pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
print(before_count)
m=SMOTE()
X_train_sm,y_train_sm=m.fit_resample(X_train,y_train)
after_counter=Counter(y_train_sm)
after_counter
```

```python
from sklearn.linear_model import LogisticRegression
LogReg=LogisticRegression()
LogReg.fit(X_train_sm,y_train_sm)
```

```python
y_predictions_smote=LogReg.predict(X_test)
y_predictions_smote
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,confusion_matrix,ConfusionMatrixDisplay
acc_log= round(accuracy_score(y_test,y_predictions_smote)*100,2)
f1_log=round(f1_score(y_test,y_predictions_smote),2)
precision_log=round(precision_score(y_test,y_predictions_smote),2)
recall_log=round(recall_score(y_test,y_predictions_smote),2)
print("accuray is:",acc_log)
print("F1 is:",f1_log)
print("Precision is:",precision_log)
print("Recall is:",recall_log)
```

```python
cmt=confusion_matrix(y_test,y_predictions_smote)

disp=ConfusionMatrixDisplay(confusion_matrix=cmt,
                            display_labels = [False, True])
disp.plot()
plt.grid(False)
plt.show()


tn, fp, fn, tp = confusion_matrix(y_test,y_predictions_smote).ravel()
print("True negative:",tn)
print("False postive:",fp)
print("False negative:",fn)
print("True postive:",tp)
```
